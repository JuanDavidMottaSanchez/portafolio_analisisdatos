# Graficos exploratorios

```python
# Distribución del tiempo en el aire
dtmaire=sns.histplot(data=datosvuelos,x='AIR_TIME')
dtmaire.set_title('Distribucion del tiempo en aire')
dtmaire.set_xlabel('Tiempo en aire')
dtmaire.set_ylabel('Cantidad de vuelos')
```

```python
## Distribución del tiempo en el aire
dtmaire=sns.boxplot(data=datosvuelos,x='AIR_TIME')
dtmaire.set_title('Distribucion del tiempo en aire')
dtmaire.set_xlabel('Tiempo en aire')
```

```python
# Distribución de la distancia
ddist=sns.kdeplot(data=datosvuelos,x='DISTANCE',fill=True)
ddist.set_title('Distribucion de la distancia recorrida')
ddist.set_xlabel('Distancia recorrida')
ddist.set_ylabel('Distribución')
```

```python
# Distribución de la distancia
ddist=sns.violinplot(data=datosvuelos,x='DISTANCE')
ddist.set_title('Distribucion de la distancia recorrida')
ddist.set_xlabel('Distancia recorrida')
ddist.set_ylabel('Distribución')
```

```python
# Comparación
ddist=sns.scatterplot(data=datosvuelos,x='DISTANCE',y='AIR_TIME')
ddist.set_title('Relacion de tiempo y distancia de vuelo')
ddist.set_xlabel('Distancia recorrida')
ddist.set_ylabel('Tiempo de vuelo')
```

```python
sns.jointplot(data=datosvuelos,x='DISTANCE',y='AIR_TIME')
```

```python
# Comparación
ddist=sns.histplot(data=datosvuelos,x='DISTANCE',y='AIR_TIME',bins=50,cbar=True)
ddist.set_title('Densidad 2D del tiempo y distancia de vuelo')
ddist.set_xlabel('Distancia recorrida')
ddist.set_ylabel('Tiempo de vuelo')
```

```python
#Agrupar los datos por mes
ddist=sns.lineplot(data=datosvuelos.groupby('MONTH').aggregate({'DISTANCE':'sum'}).reset_index(),x='MONTH',y='DISTANCE')
ddist.set_title('Total de distancia de vuelo por mes')
ddist.set_xlabel('Mes del año')
ddist.set_ylabel('Suma de distancia recorrida')
```

```python
ddist=sns.pairplot(data=datosvuelos[['DEPARTURE_TIME','DEPARTURE_DELAY','TAXI_OUT','WHEELS_OFF','SCHEDULED_TIME','ELAPSED_TIME','AIR_TIME','DISTANCE']].head(1000))
ddist.set_title('Relación de variables numericas - Primeros 100 datos')
```
